```python
import pandas as pd
'''第二次训练结果中，epoch6的准确率最高，在78.63%左右，thres=0.25'''
# 读取正确答案和预测结果的CSV文件
correct_df = pd.read_csv('cla_pre.csv')  # 包含两列 'id' 和 'label'
predicted_df = pd.read_csv('second_train/prediction6.csv')  # 包含三列 'id', 'prediction', 'label'
# my_predictions是预训练模型
# my_predictions1是16epoches训练模型

# 去掉预测结果中的文件后缀 '.jpg' 或 '.jpeg'
predicted_df['id'] = predicted_df['id'].str.replace(r'\.jpg$|\.jpeg$', '', regex=True)

# 将正确答案和预测结果通过 'id' 进行匹配
merged_df = pd.merge(predicted_df, correct_df, on='id', suffixes=('_predicted', '_correct'))
merged_df
```

```python
merged_df['Prediction'] = merged_df['score'].apply(lambda x: 1 if x > 0.5 else 0)
merged_df.to_csv('submission.csv', index=False)
```

```python
# 比较预测的结果与正确答案是否一致
merged_df['is_correct'] = merged_df['Prediction'] == merged_df['label']

# 计算正确率
accuracy = merged_df['is_correct'].mean() * 100

# 打印正确率
print(f'预测的正确率为: {accuracy:.2f}%')
merged_df = merged_df.drop(columns=['score'])
# 保存结果到新的CSV文件
merged_df.to_csv('comparison_results_21.csv', index=False)

# 打印前几行查看结果
print(merged_df.head())
```
